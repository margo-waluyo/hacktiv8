## Stamen Toner Maps

Stamen Toner is a high contrast B+W (black and white) map. Stamen Toner is perfect for data mashups and exploring river meanders and coastal zones.

```python
# create a Stamen Toner map of the world centered around Semarang
world_map = folium.Map(location=[-6.9902851, 110.4207485], zoom_start=13, tiles='Stamen Toner')

# display map
world_map
```
